```python
#Importing Library 
#Importing Dataset
import pandas as pd
import matplotlib.pylab as plt
data = pd.read_excel("https://github.com/Bhaveshpanse29/Appointment-Scheduler-project/blob/main/final_data.xlsx?raw=true")
data.info()
```

```python
#Checking for null values
data.isnull().sum()
```

```python
#Checking for Duplicate Values
data.duplicated().sum()
```

```python
#Assigning no. to columns
i=0
for a in data.columns :
  print(a+"-"+str(i))
  i = i+1
```

```python
#Droping Irrelevant Columns
ndata = data.drop(data.columns[[0,6,7,9,10,12,15,14,16,17,20,22,23,24,25]], axis = 1)
ndata.head()
```

```python
#Data filtering as per scheduled date
x = input('enter the date to get the schedule : ')
```

```python
ddata = ndata[ndata['Sample Collection Date']==x]
fdata = ddata.copy()
fdata.head()
```

```python
ddata.shape
```

```python
#Clustering data on the basis of no. of agents available
y = int(input('enter the number of agents (should be less than '+str(round(ddata.shape[0]/2))+') : '))
```

```python
#Performing label encoding for categorical data
from sklearn.preprocessing import LabelEncoder

lab  = LabelEncoder()

ddata['patient location'] = lab.fit_transform(ddata['patient location'])
ddata['Diagnostic Centers'] = lab.fit_transform(ddata['Diagnostic Centers'])
ddata['Availabilty time (Patient)'] = lab.fit_transform(ddata['Availabilty time (Patient)']) 
ddata['Test name'] = lab.fit_transform(ddata['Test name'])
ddata['Sample'] = lab.fit_transform(ddata['Sample'])
ddata.head()
```

```python
cluster = ddata[['patient location','Diagnostic Centers','shortest distance Patient-Pathlab(m)','Availabilty time (Patient)']]
cluster.head(10)
```

```python
#Performing Standardization on data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(cluster)
cluster = pd.DataFrame(scaled)
cluster.columns=['patient location','Diagnostic Centers','shortest distance Patient-Pathlab(m)','Availabilty time (Patient)']
cluster.head()
```

```python
#Using K-means Clustering to create cluster
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=y,random_state=0)

kmeans.fit(cluster)

labels = kmeans.labels_
ddata["Agent_id"]=labels
fdata["Agent_id"]=labels
ddata.head()
```

```python
#Sorting the data
z = int(input("enter the Agent_ID assigned : "))
```

```python
agent_data = fdata[fdata['Agent_id']==z]
agent_data.index = range(agent_data.shape[0])
agent_data
```

```python
t = agent_data.shape[0]
agent_data['avail']=0
for i in range(t):
  agent_data["avail"][i]=int(agent_data['Availabilty time (Patient)'][i].split('to')[0].strip().split(':')[0])

agent_data
```

```python
#Sorting agent data on shortest distance between agent and lab
Adata = agent_data.sort_values(['avail', 'shortest distance Patient-Pathlab(m)'],ascending=[1,1])
Adata
```

```python
#Data for agent for Sample Collection
first_column = Adata.pop('Availabilty time (Patient)')
Adata.insert(0, 'Availabilty time (Patient)', first_column)
Adata = Adata.drop(Adata.columns[[9]], axis = 1)
Adata
```
